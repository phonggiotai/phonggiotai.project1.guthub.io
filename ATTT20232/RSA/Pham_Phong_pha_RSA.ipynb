```python
import gmpy2
from gmpy2 import mpz

# Khai báo giá trị N cho thử thách 2
N_str = ("6484558428080716696628242653467722787263437207069762630604390703787"
         "9730861808111646271401527606141756919558732184025452065542490671989"
         "2428844841839353281972988531310511738648965962582821502504990264452"
         "1008852816733037111422964210278402893076574586452336833570778346897"
         "15838646088239640236866252211790085787877")
N = mpz(N_str)

# Bước 1: Tính giá trị gần đúng của √N
sqrt_N = gmpy2.isqrt(N)
N_1_4 = gmpy2.isqrt(gmpy2.isqrt(N))

# Bước 2: Tìm giá trị A sao cho A^2 - N là số chính phương
found = False
step = 1  # Bước nhảy 

for k in range(0, 2 * 2048 * int(N_1_4), step):
    A = sqrt_N + k
    x = A**2 - N
    
    # Bước 3: Kiểm tra x có phải là số chính phương hay không
    # Nếu x là số chính phương, thì tìm được p và q
    if x >= 0 and gmpy2.is_square(x):
        y = gmpy2.isqrt(x)
        p = A + y
        q = A - y
        found = True
        break

result = (p, q) if found else "No suitable prime factors found."
print("\n Kết quả Bài tập 2:")
print("Tìm thấy thừa số nguyên tố phù hợp:")
print(f"p = {result[0]}")
print(f"q = {result[1]}") if found else print(result)

#kiểm tra p và q có phải là số nguyên tố hay không
if gmpy2.is_prime(p) and gmpy2.is_prime(q):
    print("p và q là số nguyên tố.")

#kiểm tra tích
if p*q == N:
    print("Tích của p và q bằng N.")
```
